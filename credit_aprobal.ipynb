```python
average_income = data["Annual_income"].mean()
average_income
```

```python
gender_income = data.groupby("GENDER")["Annual_income"].mean()
gender_income
```

We can see that the males in our dataset is earinig more than the females

```python
gender_income.sort_values(ascending= False).head(1)
```

```python
sns.barplot(x=gender_income.index, y=gender_income.values, palette=["blue", "orange"])
plt.title("Average Income by Gender")
plt.xlabel("Genders")
plt.ylabel("Money")
plt.show()
```

```python
label_counts = data["label"].value_counts()
label_counts
```

```python
plt.pie(label_counts, labels=["Credit Approved", "Credit Rejected"], autopct="%1.1f%%")
plt.title("Credit Approval Status")
plt.show()
```
