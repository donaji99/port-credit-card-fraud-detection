```python
average_income = data["Annual_income"].mean()
average_income
```

```python
gender_income = data.groupby("GENDER")["Annual_income"].mean()
gender_income
```

We can see that the males in our dataset is earinig more than the females

```python
gender_income.sort_values(ascending= False).head(1)
```

```python
sns.barplot(x=gender_income.index, y=gender_income.values, palette=["blue", "orange"])
plt.title("Average Income by Gender")
plt.xlabel("Genders")
plt.ylabel("Money")
plt.show()
```

```python
label_counts = data["label"].value_counts()
label_counts
```

```python
plt.pie(label_counts, labels=["Credit Approved", "Credit Rejected"], autopct="%1.1f%%")
plt.title("Credit Approval Status")
plt.show()
```

```python
sns.boxplot(x="Marital_status", y="Annual_income", data=data, hue="Marital_status")
plt.title("Annual Income Variation by Marital Status")
plt.show()
```

```python
sns.violinplot(x= "GENDER", y = "Birthday_count", data= data, hue="GENDER")
plt.title("Distribution of birthday count by gender")
plt.show()
```

```python
data.groupby("Marital_status")["Family_Members"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Family Members Distribution by Marital Status") # Corrected title
plt.ylabel("")
plt.show()
```

```python
data.boxplot(column="Annual_income", by="EDUCATION")
plt.title("Annual Income Variation by Education Level")
plt.suptitle('') # This removes the automatic suptitle generated by `by` which can be redundant
plt.xticks(rotation=45, ha='right') # Rotate labels by 45 degrees and align them to the right
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
